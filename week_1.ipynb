```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.heatmap(data=order_df.corr())
```

```python
order_df['nama_resto'].value_counts()
```

```python
order_df['nama_resto'].value_counts().plot.bar()
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=order_df, x ='nama_resto')
plt.tight_layout()
```

```python
order_df['nama_resto'].value_counts().plot.pie(autopct='%5.1f%%', labels = None, legend=True)
plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

order_df['jumlah_brg'].plot.line(ax=ax[0][0])
ax[0][0].set_title('jumlah_brg')
```

```python
order_df.plot()
plt.tight_layout()
```

```python
order_df.hist(figsize=(6,6), bins=10)
plt.tight_layout()
```

```python
order_df.boxplot()
plt.tight_layout()
```

```python
order_df.boxplot(by="nama_resto", figsize=(8,8))
plt.tight_layout()
```

```python
sns.scatterplot(x='jumlah_brg', y='harga', data=order_df, hue='nama_resto')
plt.tight_layout()
```

```python
sns.pairplot(order_df, hue='harga', markers='+')
plt.tight_layout()
```

```python
sns.violinplot(data=order_df, y='nama_resto', x='jumlah_brg', inner='quartile')
plt.tight_layout()
```
